```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Store athlete_events in DataFrame
csv_file = "Resources/athlete_events.csv"
athlete_events_df = pd.read_csv(csv_file)
athlete_events_df.head()
```

```python
# Store noc_regions in DataFrame
csv_file = "Resources/noc_regions.csv"
noc_regions_df = pd.read_csv(csv_file)
noc_regions_df.head()
```

```python
merged_olympics = pd.merge(athlete_events_df, noc_regions_df, on="NOC", how="outer")
merged_olympics.head()
```

```python
# Rename the region and notes column to appear capitalized
merged_olympics=merged_olympics.rename(columns = {'region': 'Region', 'notes': 'Notes'})
merged_olympics
```

```python
# Create new data with select columns
new_olympics_df = merged_olympics[['Name', 'Sex', 'Age', 'Team', 'NOC', 'Year', 'Season', 'City',\
                                    'Sport', 'Medal', 'Region']].copy()
new_olympics_df
```

```python
# Clean DataFrame: remove NaN values from Medal column & remove decimal point from Year column
cleaned_olympics_df = new_olympics_df.dropna(subset=['Medal'], how='any').copy()
cleaned_olympics_df
```

```python
# def split_str_columns_df(cleaned_olympics_df,str_columns):
''' Function that splits the str columns " " is the separation, create 2 new 
    columns and remove the original. If the column's name is 'Name' the 2 new columns will be 'Name_a' and 'Name_b'.'''
# new data frame with split value columns 
# clean_olympics_df = cleaned_olympics_df
for col in ["Name"]:
#     col=str_columns[i]
    First_Name='First Name'
    Last_Name='Last Name'

    #Split
    clean_olympics_df = cleaned_olympics_df["Name"].str.rsplit(" ", n = 1, expand = True)   
    # making seperate first name column from new data frame 
    cleaned_olympics_df[First_Name]= clean_olympics_df[0]   
    # making seperate last name column from new data frame 
    cleaned_olympics_df[Last_Name]= clean_olympics_df[1] 

    # Dropping old Name columns 
    cleaned_olympics_df.drop(columns =[col], inplace = True)     

cleaned_olympics_df.head()
# return cleaned_olympics_df


```
